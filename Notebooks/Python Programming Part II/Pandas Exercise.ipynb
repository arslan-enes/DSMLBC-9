```python
df = sns.load_dataset('tips')
df.head()
```

Groupby time value and see the results of sum, min, max, mean functions

```python
df.groupby('time').agg({'total_bill':['sum','min','max','mean']})
```

```python
df.pivot_table('total_bill',['day','time'],aggfunc=['sum','min','max','mean'])
```

Lunch tips of females

```python
df.loc[np.logical_and(df['sex']=='Female',df['time']=='Lunch'),:].pivot_table(['total_bill','tip'],['day'],aggfunc=['sum','min','max','mean'])
```

Show the mean of the orders that have lower size than 3 and bigger total_bill than 10

```python
df.loc[np.logical_and(df.total_bill > 10, df['size'] < 3),:]
```

Create the total_bill_tip_sum that column will keep sum of the total_bill and tip

```python
df['total_bill_tip_sum'] = df['total_bill'] + df['tip']
```

total_bill_flag

```python
male_mean, female_mean = df.pivot_table('total_bill',['sex'],aggfunc='mean').iloc[:,0]
```

```python
male_mean
```

```python
def total_bill_flag(row):
    if row.sex == 'Female' and row.total_bill > female_mean:
        return 1
    elif row.sex == 'Male' and row.total_bill > male_mean:
        return 1
    else:
        return 0
```

```python
df['total_bill_flag'] = df.apply(total_bill_flag,axis=1)
```

male and female counts over the average bill

```python
df.groupby('sex').agg({'total_bill_flag':'sum'})
```

Sort on total_bill_tip_sum and keep first 30 value on a different dataframe

```python
new_df = df.sort_values(by='total_bill_tip_sum',ascending=False).iloc[0:30,:]
```
